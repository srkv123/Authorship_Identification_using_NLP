```python
from sklearn.model_selection import train_test_split
X=pd.read_csv('stemmedX.csv',delimiter=',')
y=df['author'].values
y.reshape(-1,1)
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(y)
y=le.transform(y)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=1/4,random_state=42)
```

```python
vocabulary = vectorizer.get_feature_names()
X_train_features=pd.DataFrame(data=X_train_trans.toarray(), columns=vocabulary)
X_train_features
```

```python
X_test_features=pd.DataFrame(data=X_test_trans.toarray(), columns=vocabulary)
X_test_features
```

```python
X_train_features['Avg words in sentence']=X_train['Avg words in sentence'].values
X_train_features['No of punctuation marks']=X_train['No of punctuation marks'].values
X_train_features['Avg sentence length']=X_train['Avg sentence length'].values
X_test_features['Avg words in sentence']=X_test['Avg words in sentence'].values
X_test_features['No of punctuation marks']=X_test['No of punctuation marks'].values
X_test_features['Avg sentence length']=X_test['Avg sentence length'].values
X_train_features
```
